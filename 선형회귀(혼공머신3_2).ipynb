```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
print(train_input)
print(train_target)

plt.scatter(train_input, train_target)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
train_dict = [[k,v] for k, v in zip(train_input,train_target)]
train_dict.sort()
train_input1 = np.array([k for [k,v] in train_dict])
train_target1 = np.array([v for [k,v] in train_dict])
train_input3 = np.array([k for [k,v] in train_dict])
train_target3 = np.array([v for [k,v] in train_dict])
```

```python
#오토인코더
#https://www.google.com/search?q=%EC%98%A4%ED%86%A0%EC%9D%B8%EC%BD%94%EB%8D%94&oq=%EC%98%A4%ED%86%A0%EC%9D%B8%EC%BD%94%EB%8D%94&aqs=chrome..69i57.2087j0j4&sourceid=chrome&ie=UTF-8

for t in range(30):
  train_dict = [[k,v] for k, v in zip(train_input1,train_target1)]
  train_dict.sort()
  train_input1 = np.array([k for [k,v] in train_dict])
  train_target1 = np.array([v for [k,v] in train_dict])

  train_input2 = np.array([])
  train_target2 = np.array([])

  for i in range(len(train_input1)-8):
    x1 = train_input1[i]
    x2 = train_input1[i+8]
    y1 = train_target1[i]
    y2 = train_target1[i+8]
    train_input2 = np.append(train_input2,(x1+x2)/2)
    train_target2 = np.append(train_target2,(y1+y2)/2)


  train_input3 = np.append(train_input3,train_input2)
  train_target3 = np.append(train_target3,train_target2)
  train_input1 = train_input2
  train_target1 = train_target2
print(train_input3)
print(train_target3)
```

```python
plt.scatter(train_input3, train_target3)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
train_input3 = train_input3.reshape(-1,1)
test_input3 = test_input3.reshape(-1,1)
train_input = train_input.reshape(-1,1)
test_input = test_input.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(train_input3, train_target3)
linear_predict = lr.predict([[50]])
print(linear_predict)
print(lr.coef_, lr.intercept_)

plt.scatter(train_input3, train_target3)
plt.plot([15, 50],[15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])
# 내가 찾는 값은 50의 무게를 가진 농어의 길이이다.
# 그렇기 때문에 15 ~ 50의 무게에 대한 1차방정식 그래프를 그린다. [15,50]
# 다음 값은 15*w + b 로써, y값을 의미한다.

plt.scatter(50, linear_predict, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show

print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))

"""
기존 결정계수의 점수 값
0.9398463339976039
0.8247503123313558
"""
```

```python
train_poly3 = np.column_stack((train_input3 ** 2, train_input3))
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input ** 2, test_input))
lr.fit(train_poly3, train_target3)
multi_predict = lr.predict([[80**2,80]]) # 80cm 농어 약 4~5kg
multi_predict2 = lr.predict([[70**2,70]]) # 70cm 농어 약 3.3kg ~ 3.6kg
multi_predict3 = lr.predict([[60**2,60]]) # 60cm 농어 약 2kg
multi_predict4 = lr.predict([[105.3**2,105.3]]) # 105.3cm 농어 약 9.29kg , 7.56, 
multi_predict5 = lr.predict([[125**2,125]]) # 125cm 농어 12.88kg
print(multi_predict)
print(multi_predict2)
print(multi_predict3)
print(multi_predict4)
print(multi_predict5)
print(lr.coef_, lr.intercept_)


point = np.arange(15,126) # 조금더 그래프를 늘림
plt.scatter(train_input3, train_target3)
plt.plot(point, lr.coef_[0]*point**2 + lr.coef_[1]*point + lr.intercept_)
plt.scatter(80, 4500, marker='^')
plt.scatter(70, 3300, marker='^')
plt.scatter(60, 2000, marker='^')
plt.scatter(105.3, 7560, marker='^')
plt.scatter(105.3, 9290, marker='^')
plt.scatter(125, 12880, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show

#결정계수 R**2 확인
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))

"""
기존 결정계수 값
0.9706807451768623
0.9775935108325122
"""
```
